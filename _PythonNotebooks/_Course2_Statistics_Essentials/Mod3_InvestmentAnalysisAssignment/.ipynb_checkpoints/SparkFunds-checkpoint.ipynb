```python
#Assignment Statistics Essentials - Investment Analysis Assignment
import numpy as np
import pandas as pd
import ntpath
from datetime import datetime
from  matplotlib.ticker import FuncFormatter
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Opening the companies.csv in Excel says that there is a circular reference in R46673C2.
#This could be a typo error in the name "=-R- Ranch and Mine" which is interpreted as a formula by Excel
compFile = "companies.csv"
roundsFile = "rounds2.csv"
mappingsFile = "mapping.csv"

#Encoding error seen - fix it by mentioning the engine name as 'python'
#Step 1. Open the companies and print their info to know NULLs in columns that we want to analyze
companies = pd.read_csv(compFile, engine='python')
rounds = pd.read_csv(roundsFile, engine='python')
mappings = pd.read_csv(mappingsFile, engine='python')

#Print info
print("Companies permalink not null for all 66368 entries")
print("**************************************************")
print(companies.info())
print("Rounds info funding_round_permalink not null for all 114949 entries")
print("*******************************************************************")
print(rounds.info())
print("Category mapping info entries have 687 of 688 not null values - One is blank")
print("*****************************************************************************")
print(mappings.info())
print("Checking if every category is mapped to exactly ONE sector")
print("***********************************************************")

def checkCatMapped2OneSector(mp):
    onesColumn = mp.loc[:, mp.columns != 'category_list']
    onesColumn['justOne'] = onesColumn.sum(axis=1)

    checkFor1 = lambda x : True if x == 1 else False
    #Mean or count will not be accurate to say if every entry is exactly having 1.
    #onesColumn['justOne'].mean() or onesColumn['justOne'].count() will not accurately tell this
    all1s = all(onesColumn['justOne'].apply(checkFor1))
    if all1s == True:
        print("************Mapping data okay************")
    else:
        print("************Mapping data NOT okay************")
        print("************REPORT ERROR - A category assigned to more than 1 Sector************")
    
    return

checkCatMapped2OneSector(mappings)
```

###### Section 5
-  removeRaisedAmtUSDOutliers - On the given dataFrame, for a given investment type, remove outliers beyond p1 and p2 exclusive. Calculate the Median and boxplot to know where the investment type stands.
-  removeOutliersForEACHRoundTypeAnd4BoxPlots - Call removeRaisedAmtUSDOutliers for the 4 investment types of interest.
-  removeOutliersForEACHRoundTypeAndBoxPlotMerged - Uses removeRaisedAmtUSDOutliers and finally concats the 4 dataframes and does 1 boxplot. The boxplot is squished and is not readable. Use removeOutliersForEACHRoundTypeAnd4BoxPlots instead.

```python
from  matplotlib.ticker import FuncFormatter
#For the investment type, get the date between p1 and p2 percentiles and boxplot
def removeRaisedAmtUSDOutliers(dataFrame, rndType, p1, p2):
    print("")
    print("****************************************************")
    print('BOX PLOTTING {} BETWEEN {} AND {} percentile'.format(rndType, p1, p2))
    print("****************************************************")
    if p2 < 1 or p1 > p2:
        print("Error - p1, p2 should be between 1 and 99, p1 should be less than p2")
        return None
    
    rndType = rndType.upper()

    rndFltr = dataFrame['funding_round_type_upper'].isin([rndType])
    dataFrame = dataFrame[rndFltr]
    
    np1 = np.percentile(dataFrame['raised_amount_usd'], p1)
    print(p1, " percentile of investment amount " , np1)        
    np2 = np.percentile(dataFrame['raised_amount_usd'], p2)
    print(p2, " percentile of investment amount " , np2)
    
    percfltr = (np1 <= dataFrame['raised_amount_usd']) & (dataFrame['raised_amount_usd'] <= np2)
    dataFrame = dataFrame.loc[percfltr]
    med = np.median(dataFrame['raised_amount_usd'])
    print("Median is ", med)
    
    plt.figure(figsize=[8,12])
    plt.boxplot(dataFrame.raised_amount_usd)
    plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: '${:0,.0f}'.format(x)))
    plt.show()
    
    return dataFrame

#Outliers are removed per funding type and median is calculated.
#BoxPlots all the 4 Funding types in 4 separate charts.
def removeOutliersForEACHRoundTypeAnd4BoxPlots(dataFrame, p1, p2):    
    rndTypeLst = ['ANGEL', 'SEED', 'VENTURE', 'PRIVATE_EQUITY']
    
    for rndType in rndTypeLst:
        removeRaisedAmtUSDOutliers(dataFrame, rndType, p1, p2)
    
    return

#This method will give a squished, picture on a uniform scale, 
#difficult to read. IGNORE THIS METHOD
#Outliers are removed per funding type and median is calculated.
#BoxPlots all the 4 Funding types in ONE single chart.
def removeOutliersForEACHRoundTypeAndBoxPlotMerged(dataFrame, p1, p2):
    rndTypeLst = ['ANGEL', 'SEED', 'VENTURE', 'PRIVATE_EQUITY']
    
    df1 = removeRaisedAmtUSDOutliers(dataFrame, 'ANGEL', p1, p2)
    df2 = removeRaisedAmtUSDOutliers(dataFrame, 'SEED', p1, p2)
    df3 = removeRaisedAmtUSDOutliers(dataFrame, 'VENTURE', p1, p2)
    df4 = removeRaisedAmtUSDOutliers(dataFrame, 'PRIVATE_EQUITY', p1, p2)
    
    unionDataFrame = pd.concat([df1, df2, df3, df4])
    print(unionDataFrame.info())
    
    plt.figure(figsize=[6,8])
    sns.boxplot(unionDataFrame['funding_round_type_upper'], unionDataFrame['raised_amount_usd'])
    plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: '${:0,.0f}'.format(x)))
    plt.show()

    return


#['ANGEL', 'SEED', 'VENTURE', 'PRIVATE_EQUITY']
#removeOutliersForEACHRoundTypeAndBoxPlotMerged(master_frame, 5, 95)
#removeOutliersForEACHRoundTypeAndBoxPlotMerged(master_frame, 25, 75)
#removeOutliersForEACHRoundTypeAndBoxPlotMerged(master_frame, 0, 95)
```

##### Section 8
-  Apply the business rule to extract the primary_sector from category_list in companies.
-  Identify the categories without mappings.
-  Use 'fixSpelling' to fix spelling mistakes and redo the previous step to check if the categories without mappings are reduced. The pattern is simple where only 52 entries have 0 inplace of 'na'.
-  The final list of categories without mappings will be treated as 'blanks'.
-  'mainSectorFromPrimarySector' - For a given 'row' in the category mappings dataframe, it checks for '1' in a column and returns that as the 'main_sector'
-  Apply 'mainSectorFromPrimarySector' and get the 'main_sector'

```python
#Checkpoint 4: Sector Analysis 1
print("****************CHECKPOINT 4: Sector Analysis 1****************")

primarySector = lambda pipeSeperatedList : pipeSeperatedList.split('|')[0]

#Pick the category before first pipe as the 'primary sector'
master_frame['primary_sector'] = master_frame['category_list_lower'].apply(primarySector)

#Get unique of the primary_sector & mapping category print their counts.
companyPS = set(master_frame['primary_sector'])
print("Count of unique Primary sect ", len(companyPS))
mappingCL = set(mappings['category_list_lower'])
print("Count of unique mapping CL ", len(mappingCL))

#Get the company category names that don't have a mapping defined in mapping file.
#This is done by companyPS.difference(mappingCL).
#If this list is non-empty, it means there are missing mappings (that will be mapped 
#to blank_unknown) or there are spelling errors in either columns.

catWithoutMapping = sorted(companyPS.difference(mappingCL))
print("**************************************************")
print("*****************Count of catWithoutMapping ", len(catWithoutMapping), " *********")
print("*****************Category with out mapping in companies*********")
print(catWithoutMapping)

print("\n\n********************************************************************")
print("IMPORTANT: Skimming through the above list we see NO spelling mistakes which implies\n",
      "there are spell errors in mapping['category_list_lower'] OR missing mappings that\n",
      "should be mapped to 'blank_unknown'. NOTE that we are fixing in mapping['category_list_lower'].")
print("********************************************************************\n")

print("*****************************************************************")
print("Skimming through the 688 entries of mapping['category_list_lower']")
print("the spelling mistakes are due to alphabet sequence 'na' replaced by")
print("numeral 0. Find replace 0 with 'na' works except for Enterprise 2.0.")
print("Applying a filter in Excel, there are 52 entries that need to be fixed.")
print("*****************************************************************")

print("***********************************************")

def fixSpelling(category) : 
    if not category.endswith('.0'):
        return category.replace('0', 'na') 
    return category

mappings['primary_sector'] = mappings['category_list_lower'].apply(fixSpelling)

#CHECK if the data is fixed by doing companyPS.difference(newSet) to check if we have only 37 entries left
newSet = sorted(set(mappings['primary_sector']))
v2CatWithoutMapping = sorted(companyPS.difference(newSet))
print("**************************************************")
print("******REDUCED Count of v2CatWithoutMapping ", len(v2CatWithoutMapping), " ******")
print("***********************************************")
print(v2CatWithoutMapping)
```

```python
#Checkpoint 4 - Sector Analysis 1
#Map the primary sector extracted from the pipe separated category list to the main_sector
#Code for a merged data frame with each primary sector mapped to its main sector 
#(the primary sector should be present in a separate column).

def mainSectorFromPrimarySector(row):
    #Every item in row.items is a tuple
    #that can be accessed by item[0] and item[1] or 
    #with a pair of variables
    for index, value in row.items():
        if (index != 'primary_sector' and value == 1):
            #print("Key - ", index, " Value - ", value)
            return index.upper()
    #We will never reach here because we have already checked if the 
    #rows have exactly one numeric value '1' in the first section of this notebook
    return "BLANKS" #Equivalent of no mapping.

#Pick only the 9 SECTOR COLUMNS, including BLANKS _with_ 'primary_sector'
columnFilter = ~mappings.columns.isin(['category_list', 'category_list_lower'])
sector_mapping_frame = mappings.loc[:, columnFilter]
sector_mapping_frame['main_sector'] = sector_mapping_frame.apply(mainSectorFromPrimarySector, axis=1)
#primary_sector and main_sector mapped in sector_mapping_frame

#PRINT TO CHECK
#pd.set_option('display.max_rows', 100)
print(sector_mapping_frame[['primary_sector', 'main_sector']])
#pd.set_option('display.max_rows', 10)
```

##### Section 9
-  Merge the main_sector to master_frame for sector analysis.
-  'sectrMappedCosByCountry' - Takes the sector mapped companies and returns the dataframe for the Country, applying the filters VENTURE type and Funding amount consideration of 5M to 15M.
- Use 'sectrMappedCosByCountry' and build the 3 dataframes D1, D2, and D3 for the top 3 English speaking countries. 

```python
#Checkpoint 5 - Sector Analysis 2
#Merge the sector_mapping_frame with main_dataframe by 'primary_sector'
#The 'master_frame' is merged with 'sector_mapping_frame' by the 'primary_sector'
#Since we derived the 'main_sector', we just need the 2 columns 'primary_sector' and 'main_sector'
sector_mapped_companies = pd.merge(master_frame, \
                                   sector_mapping_frame[['primary_sector', 'main_sector']], 
                                   how='inner', on='primary_sector')

print("***************************************************")
print("Number of companies that don't have sector mapping - ", len(master_frame) - len(sector_mapped_companies))
print("***************************************************")
print(sector_mapped_companies.columns)

#Create three separate data frames D1, D2 and D3 for each of the three countries('USA', 'GBR' and 'IND') containing the 
#observations of funding type FUNDING_TYPE falling within the 5-15 million USD range. 

def sectrMappedCosByCountry(smc, country):
    F5M = 5000000
    F15M = 15000000
    
    DFilter = smc['country_code'].isin([country]) & \
            smc['funding_round_type_upper'].isin([FUNDING_TYPE]) & \
            (F5M <= smc['raised_amount_usd']) & \
            (smc['raised_amount_usd'] <= F15M)
    D = smc[DFilter]
    
    return D

D1 = sectrMappedCosByCountry(sector_mapped_companies, 'USA')
D2 = sectrMappedCosByCountry(sector_mapped_companies, 'GBR')
D3 = sectrMappedCosByCountry(sector_mapped_companies, 'IND')
```

##### Section 10
-  'highstInvCompanyInSect' - For a given sector, finds the company that has got the highest investment amount from the dataframe that is already filtered by Country and Investment Type.
-  'sectorWiseAnalyze' - For the dataframes like D1, D2 & D3 (already filtered by Country), find the sector-wise investment count details. After this, calls 'highstInvCompanyInSect' to print company with highest investment amount in the top sector.
- Call 'sectorWiseAnalyze' for the 3 countries.

```python
def highstInvCompanyInSect(invstDF, sector):
    highInvCoInSectorFltr = invstDF['main_sector'].isin([sector])
    indx = ["company_permalink_lower", "name"]
    
    highInvCoInSector = pd.pivot_table(data=invstDF[highInvCoInSectorFltr], index=indx, \
                            values="raised_amount_usd", aggfunc = np.sum).\
                            reset_index().rename(columns={'raised_amount_usd': 'Investment Amount'})
    highInvCoInSector = highInvCoInSector.sort_values(by=['Investment Amount'], ascending=False)
        
    print('FOR THE TOP SECTOR \"{}\" COUNT-WISE, WHICH COMPANY \nRECEIVED THE HIGHEST INVESTMENT?'.format(sector))
    print("***************************************************")
    
    highInvCoInSector = highInvCoInSector.reset_index(drop=True)
    w = highInvCoInSector.iloc[0]['company_permalink_lower']
    n = highInvCoInSector.iloc[0]['name']
    amt = highInvCoInSector.iloc[0]['Investment Amount']
    print('COMPANY WEBLINK - {}, COMPANY NAME - {}, \n\
    INVESTMENT AMOUNT - ${:0,.0f}'.format(w, n, amt))
    
    return

def sectorWiseAnalyze(invDF, country):
    print("\n\n")
    print("***************************************************")
    print("Sector-wise Investment details for ", country)
    print("***************************************************")
    print('1) Total number of investments (count) {:0,.0f}'.format(len(invDF)))
    
    sumAmt = invDF['raised_amount_usd'].sum()
    print('2) Total amount of investment (USD) ${:0,.0f}'.format(sumAmt))
    
    grpSec = pd.pivot_table(data=invDF, index="main_sector", values="raised_amount_usd", aggfunc = 'count').\
                reset_index().rename(columns={'raised_amount_usd': 'Number of Investments'})
    
    grpSec = grpSec.sort_values(by=['Number of Investments'], ascending=False)

    #print("********Ordered sector-wise count of investments*****************")
    #print(grpSec)
    #print("\n")

    topSecCnt = grpSec[:3]
    
    topSecCntIndexable =  topSecCnt.reset_index(drop=True)
    print("")
    print("3, 4, 5, 6, 7, 8) TOP 3 SECTORS (based on count of investments)")
    print(topSecCntIndexable)
    print("")
    tS1 = topSecCntIndexable.iloc[0]['main_sector']
    tS2 = topSecCntIndexable.iloc[1]['main_sector']
    print("TOP 1st SECTOR BY COUNT IS - ", tS1)
    print("TOP 2nd SECTOR BY COUNT IS - ", tS2)
    
    print("9)")
    highstInvCompanyInSect(invDF, tS1)
    print("10)")
    highstInvCompanyInSect(invDF, tS2)
    
    return topSecCnt

#USE THESE FOR PLOTTING LATER
c1 = sectorWiseAnalyze(D1, "USA")
c2 = sectorWiseAnalyze(D2, "GBR")
c3 = sectorWiseAnalyze(D3, "IND")
```
